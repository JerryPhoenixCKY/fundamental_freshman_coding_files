# importing necessary libraries

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

# importing Breast Cancer Dataset from sklearn. 

The features of breast cancer tumors are included in this dataset, along with labels that correlate to them: 0 for malignant and 1 for benign

```python
# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target
print(X.shape)
print(y.shape)
```

# splitting the dataset into train and test set

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# creating a Perceptron model

```python
# Create a Perceptron model
clf = Perceptron(max_iter=1000, eta0=0.1) # max iteration count and learning rate
```

# Train the Perceptron model

```python
# Train the model
clf.fit(X_train, y_train)
```

# use trained Perceptron model to make predictions on test data after training

```python
# Make predictions
y_pred = clf.predict(X_test)
print(y_pred)
```

# Evaluation: assess the model's accuracy in identifying malignant or benign tumors

```python
# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

# confusion matrix

```python
# Calculate confusion matrix
import matplotlib.pyplot as plt
conf_matrix = confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = conf_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```

# Beyond Binary Classification:Multiclass Classification for handwritten digits

# load necessary libs, including the MNIST dataset from sklearn.datasets

```python
from sklearn.datasets import fetch_openml
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns
```

use the fetch_openml method from sklearn to import the MNIST dataset, which comprises of handritten digits (0 - 9) in 28x28 pixel pictures and their related labels

```python
# Load the MNIST dataset
mnist = fetch_openml('mnist_784')
X = mnist.data
y = mnist.target.astype(int)
print(X.shape)
print(y.shape)
```

split the dataset into train and test set, and create a perceptron model

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Create a Perceptron model
perceptron = Perceptron(max_iter=1000, eta0=0.1)
```

model training and prediction

```python
# Train the model
perceptron.fit(X_train, y_train)
# Make predictions
y_pred = perceptron.predict(X_test)
print(y_pred)
```

evaluation: determine the accuracy of the model's prediction to evaluate its performance

```python
# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

confusion matrix

```python
# Calculate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

sns.heatmap(conf_matrix, annot = True, cmap= 'Blues')
plt.ylabel('True')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
```
