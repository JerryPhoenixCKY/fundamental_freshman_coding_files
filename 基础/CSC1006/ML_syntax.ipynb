 Step 1: Import Required Libraries

```python
# 导入numpy用于数值计算
# Import numpy for numerical computations
import numpy as np

# 导入matplotlib用于数据可视化
# Import matplotlib for data visualization
import matplotlib.pyplot as plt

# 从sklearn导入数据集模块
# Import datasets module from sklearn
from sklearn import datasets

# 从sklearn导入模型选择模块（用于数据划分）
# Import model_selection module from sklearn (for data splitting)
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV

# 从sklearn导入预处理模块（用于数据标准化）
# Import preprocessing module from sklearn (for data standardization)
from sklearn.preprocessing import StandardScaler

# 从sklearn导入多种分类算法
# Import multiple classification algorithms from sklearn
from sklearn.svm import SVC  # 支持向量机 / Support Vector Machine
from sklearn.neighbors import KNeighborsClassifier  # K近邻 / K-Nearest Neighbors
from sklearn.tree import DecisionTreeClassifier  # 决策树 / Decision Tree
from sklearn.ensemble import RandomForestClassifier  # 随机森林 / Random Forest
from sklearn.neural_network import MLPClassifier  # 神经网络 / Neural Network
from sklearn.linear_model import LogisticRegression  # 逻辑回归 / Logistic Regression

# 从sklearn导入评估指标
# Import evaluation metrics from sklearn
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import precision_recall_fscore_support

# 从sklearn导入管道模块（用于构建处理流程）
# Import pipeline module from sklearn (for building processing pipelines)
from sklearn.pipeline import Pipeline

# 导入joblib用于模型保存和加载
# Import joblib for model saving and loading
import joblib

# 设置随机种子以保证结果可重现
# Set random seed for reproducible results
np.random.seed(42)
```

Step 2:  Load and Explore Data

```python
# 加载手写数字数据集（MNIST的简化版本，8x8像素）
# Load handwritten digits dataset (simplified version of MNIST, 8x8 pixels)
digits = datasets.load_digits()

# 查看数据集的基本信息
# View basic information about the dataset
print("数据集描述 / Dataset description:")
print(f"数据形状 / Data shape: {digits.data.shape}")  # (1797, 64) - 1797个样本，64个特征
print(f"标签形状 / Target shape: {digits.target.shape}")  # (1797,) - 1797个标签
print(f"特征名称 / Feature names: {digits.feature_names}")  # 像素值名称 / pixel value names
print(f"类别名称 / Target names: {digits.target_names}")  # 0-9十个数字 / digits 0-9
print(f"图像形状 / Image shape: {digits.images[0].shape}")  # (8, 8) - 每个图像是8x8像素

# 显示数据集的基本统计信息
# Display basic statistics of the dataset
print(f"\n数据统计 / Data statistics:")
print(f"特征最小值 / Min value: {digits.data.min()}")
print(f"特征最大值 / Max value: {digits.data.max()}")
print(f"特征均值 / Mean value: {digits.data.mean():.2f}")
print(f"特征标准差 / Std value: {digits.data.std():.2f}")

# 显示类别分布
# Show class distribution
unique, counts = np.unique(digits.target, return_counts=True)
print(f"\n类别分布 / Class distribution:")
for digit, count in zip(unique, counts):
    print(f"数字 {digit}: {count} 个样本 / samples")
```

Step 3: Data Visualization

```python
# 创建一个图形来显示前10个手写数字图像
# Create a figure to display the first 10 handwritten digit images
fig, axes = plt.subplots(2, 5, figsize=(12, 6))
axes = axes.ravel()  # 将2x5的网格展平为一维数组 / Flatten 2x5 grid to 1D array

# 循环显示前10个图像
# Loop to display the first 10 images
for i in range(10):
    # 将数据重塑为8x8的图像
    # Reshape the data to 8x8 image
    axes[i].imshow(digits.images[i], cmap='gray')
    # 设置标题显示真实的数字标签
    # Set title showing the true digit label
    axes[i].set_title(f'数字: {digits.target[i]}')
    # 隐藏坐标轴
    # Hide axes
    axes[i].axis('off')

plt.tight_layout()  # 调整子图间距 / Adjust subplot spacing
plt.show()

# 显示图像的像素值分布
# Show pixel value distribution of images
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(digits.data.ravel(), bins=20, edgecolor='black')
plt.title('像素值分布 / Pixel Value Distribution')
plt.xlabel('像素值 / Pixel Value')
plt.ylabel('频数 / Frequency')

plt.subplot(1, 2, 2)
plt.boxplot(digits.data)
plt.title('特征值箱线图 / Feature Value Boxplot')
plt.xlabel('特征维度 / Feature Dimension')
plt.ylabel('像素值 / Pixel Value')
plt.tight_layout()
plt.show()
```

Step 4: Data Preprocessing

```python
# 提取特征和标签
# Extract features and labels
X = digits.data  # 特征矩阵 / Feature matrix (1797, 64)
y = digits.target  # 标签 / Labels (1797,)

# 划分训练集和测试集（70%训练，30%测试）
# Split into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.3,  # 30%用于测试 / 30% for testing
    random_state=42,  # 随机种子保证可重现 / Random seed for reproducibility
    stratify=y  # 保持类别比例 / Maintain class proportions
)

print(f"训练集大小 / Training set size: {X_train.shape}")
print(f"测试集大小 / Testing set size: {X_test.shape}")
print(f"训练集类别分布 / Training set class distribution:")
print(np.bincount(y_train))
print(f"测试集类别分布 / Testing set class distribution:")
print(np.bincount(y_test))

# 创建标准化器（将特征转换为均值为0，标准差为1）
# Create scaler (transform features to mean=0, std=1)
scaler = StandardScaler()

# 对训练集进行拟合和转换
# Fit and transform the training set
X_train_scaled = scaler.fit_transform(X_train)

# 对测试集进行转换（使用训练集的均值和标准差）
# Transform the test set (using training set mean and std)
X_test_scaled = scaler.transform(X_test)

print(f"\n标准化后的训练集统计 / Training set statistics after scaling:")
print(f"均值 / Mean: {X_train_scaled.mean():.10f}")  # 应该接近0 / Should be close to 0
print(f"标准差 / Std: {X_train_scaled.std():.2f}")   # 应该接近1 / Should be close to 1
```

Step 5: Create and Train Multiple Models

<img src="https://github.com/user-attachments/assets/95127c00-b49b-4fde-b0d2-faec93d5ad06" width=1000>

```python
# 定义要比较的模型
# Define models to compare
models = {
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=3),
    'Decision Tree': DecisionTreeClassifier(max_depth=10, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'SVM (RBF)': SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42),
    'SVM (Linear)': SVC(kernel='linear', C=1.0, random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Neural Network': MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
}

# 训练和评估每个模型
# Train and evaluate each model
results = {}
for name, model in models.items():
    print(f"\n训练模型: {name} / Training model: {name}")
    
    # 训练模型
    # Train the model
    model.fit(X_train_scaled, y_train)
    
    # 在测试集上进行预测
    # Make predictions on test set
    y_pred = model.predict(X_test_scaled)
    
    # 计算准确率
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy
    
    print(f"测试集准确率 / Test accuracy: {accuracy:.4f}")
    
    # 计算精确率、召回率和F1分数
    # Calculate precision, recall and F1 score
    precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='weighted')
    print(f"精确率 / Precision: {precision:.4f}")
    print(f"召回率 / Recall: {recall:.4f}")
    print(f"F1分数 / F1 Score: {f1:.4f}")

# 显示所有模型的比较结果
# Display comparison results of all models
print("\n" + "="*50)
print("模型比较结果 / Model Comparison Results:")
print("="*50)
for name, accuracy in sorted(results.items(), key=lambda x: x[1], reverse=True):
    print(f"{name:20s}: {accuracy:.4f}")
```

Step 6: Hyperparameter Tuning

```python
# 选择最佳模型（这里以SVM为例）
# Select the best model (SVM as example)
print("对SVM模型进行超参数调优 / Hyperparameter tuning for SVM model")

# 定义要搜索的参数网格
# Define parameter grid for search
param_grid = {
    'C': [0.1, 1, 10, 100],  # SVM的正则化参数 / SVM regularization parameter
    'gamma': [0.001, 0.01, 0.1, 1],  # SVM核函数系数 / SVM kernel coefficient
    'kernel': ['rbf', 'linear']  # SVM核函数类型 / SVM kernel type
}

# 创建SVM模型（未标准化，因为我们会在调优前对数据进行标准化）
# Create SVM model (not scaled, because we will scale data before tuning)
svm_model = SVC(random_state=42)

# 创建网格搜索对象（使用5折交叉验证）
# Create GridSearchCV object (using 5-fold cross-validation)
grid_search = GridSearchCV(
    svm_model,  # 要调优的模型 / Model to tune
    param_grid,  # 参数网格 / Parameter grid
    cv=5,  # 5折交叉验证 / 5-fold cross-validation
    scoring='accuracy',  # 评分标准 / Scoring metric
    n_jobs=-1,  # 使用所有CPU核心 / Use all CPU cores
    verbose=1  # 显示进度 / Show progress
)

# 执行网格搜索（使用标准化后的数据）
# Perform grid search (using scaled data)
print("开始网格搜索 / Starting grid search...")
grid_search.fit(X_train_scaled, y_train)

print(f"\n最佳参数 / Best parameters: {grid_search.best_params_}")
print(f"最佳交叉验证分数 / Best cross-validation score: {grid_search.best_score_:.4f}")

# 使用最佳模型进行预测
# Make predictions with best model
best_svm = grid_search.best_estimator_
y_pred_best = best_svm.predict(X_test_scaled)

# 评估最佳模型
# Evaluate best model
accuracy_best = accuracy_score(y_test, y_pred_best)
print(f"最佳模型测试集准确率 / Best model test accuracy: {accuracy_best:.4f}")
```
