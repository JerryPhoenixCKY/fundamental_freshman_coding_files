```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title("Sine Wave")
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load data & split
X, y = load_iris(return_X_y=True)
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3, random_state=0)

# Online learner (learns step-by-step)
model = SGDClassifier(loss="log_loss", max_iter=1, learning_rate="optimal", tol=None)

print("Training...\n")
for epoch in range(1, 11):
    model.partial_fit(Xtr, ytr, classes=[0,1,2])
    acc = accuracy_score(yte, model.predict(Xte))
    print(f"Epoch {epoch:02d} | Test accuracy: {acc:.3f}")

print("\nDone! Final prediction:", model.predict([Xte[0]]))
```
