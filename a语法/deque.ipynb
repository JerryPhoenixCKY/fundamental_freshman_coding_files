### 2. åŸºæœ¬æ“ä½œï¼šæ·»åŠ å’Œåˆ é™¤

```python
from collections import deque

d = deque()
d.append(1)        # å³åŠ 
d.append(2)
d.appendleft(0)    # å·¦åŠ 
print(d)           # deque([0, 1, 2])

x = d.popleft()    # å·¦åˆ  â†’ x = 0
y = d.pop()        # å³åˆ  â†’ y = 2
print("åˆ é™¤:", x, y)  # åˆ é™¤: 0 2
print(d)           # deque([1])
```

```python
from collections import deque

# æœ€å¤šå­˜3ä¸ªå…ƒç´ 
d = deque(maxlen=3)
d.append(1)
d.append(2)
d.append(3)
print(d)  # deque([1, 2, 3], maxlen=3)

d.append(4)  # è¶…å‡ºé•¿åº¦ï¼Œè‡ªåŠ¨ä»å·¦è¾¹åˆ æ‰1
print(d)     # deque([2, 3, 4], maxlen=3)

d.appendleft(0)  # ä»å·¦è¾¹åŠ 0ï¼Œå³è¾¹4è¢«æŒ¤æ‰
print(d)         # deque([0, 2, 3], maxlen=3)
```

### 2. å…¶ä»–å®ç”¨æ–¹æ³•

| æ–¹æ³• | ä½œç”¨ |
|------|------|
| `extend(iterable)` | ä»å³è¾¹æ‰¹é‡æ·»åŠ å¤šä¸ªå…ƒç´  |
| `extendleft(iterable)` | ä»å·¦è¾¹æ‰¹é‡æ·»åŠ ï¼ˆæ³¨æ„ï¼šé¡ºåºä¼šåè½¬ï¼ï¼‰ |

#### ğŸ’¡ ç¤ºä¾‹ï¼š

```python
d = deque([1, 2, 3])
d.extend([4, 5])        # å³åŠ å¤šä¸ª
print(d)                # deque([1, 2, 3, 4, 5])

d.extendleft([0, -1])   # å·¦åŠ å¤šä¸ªï¼Œæ³¨æ„é¡ºåºåè½¬ï¼
print(d)                # deque([-1, 0, 1, 2, 3, 4, 5])

d.rotate(2)             # å‘å³æ—‹è½¬2ä½
print(d)                # deque([4, 5, -1, 0, 1, 2, 3])

d.rotate(-1)            # å‘å·¦æ—‹è½¬1ä½
print(d)                # deque([5, -1, 0, 1, 2, 3, 4])
```

## ğŸ§  å››ã€é«˜çº§åº”ç”¨åœºæ™¯ï¼ˆèµ„æ·±å¼€å‘è€…ï¼‰

### åœºæ™¯1ï¼šé«˜æ•ˆå®ç°é˜Ÿåˆ—ï¼ˆQueueï¼‰

Python çš„ `list` ç”¨ä½œé˜Ÿåˆ—æ—¶ï¼Œ`pop(0)` æ˜¯ O(n)ï¼Œæ•ˆç‡ä½ã€‚
è€Œ `deque` çš„ `popleft()` æ˜¯ O(1)ï¼Œæ˜¯**ç†æƒ³çš„é˜Ÿåˆ—å®ç°**ã€‚

```python
from collections import deque

queue = deque()
queue.append("ä»»åŠ¡1")
queue.append("ä»»åŠ¡2")
queue.append("ä»»åŠ¡3")

while queue:
    task = queue.popleft()  # é«˜æ•ˆå–å‡ºé˜Ÿé¦–ä»»åŠ¡
    print("å¤„ç†:", task)
```

### åœºæ™¯2ï¼šå¹¿åº¦ä¼˜å…ˆæœç´¢ï¼ˆBFSï¼‰

åœ¨å›¾æˆ–æ ‘çš„ BFS ä¸­ï¼Œéœ€è¦ç”¨é˜Ÿåˆ—å­˜å‚¨å¾…è®¿é—®èŠ‚ç‚¹ã€‚

```python
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node)
            visited.add(node)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)

# ç¤ºä¾‹å›¾
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
bfs(graph, 'A')
```

### åœºæ™¯3ï¼šæ»‘åŠ¨çª—å£æœ€å¤§å€¼ï¼ˆç»å…¸ç®—æ³•é¢˜ï¼‰

ç»™å®šæ•°ç»„å’Œçª—å£å¤§å° kï¼Œæ±‚æ¯ä¸ªçª—å£å†…çš„æœ€å¤§å€¼ã€‚

```python
from collections import deque

def max_sliding_window(nums, k):
    if not nums or k == 0:
        return []

    dq = deque()  # å­˜å‚¨ç´¢å¼•ï¼Œä¿è¯é˜Ÿé¦–æ˜¯å½“å‰çª—å£æœ€å¤§å€¼çš„ç´¢å¼•
    result = []

    for i in range(len(nums)):
        # ç§»é™¤è¶…å‡ºçª—å£çš„ç´¢å¼•
        if dq and dq[0] < i - k + 1:
            dq.popleft()

        # ç»´æŠ¤å•è°ƒé€’å‡é˜Ÿåˆ—ï¼šç§»é™¤æ¯”å½“å‰å…ƒç´ å°çš„å°¾éƒ¨å…ƒç´ 
        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()

        dq.append(i)

        # çª—å£å½¢æˆåå¼€å§‹è®°å½•ç»“æœ
        if i >= k - 1:
            result.append(nums[dq[0]])

    return result

# æµ‹è¯•
print(max_sliding_window([1,3,-1,-3,5,3,6,7], 3))
# è¾“å‡º: [3, 3, 5, 5, 6, 7]
```
