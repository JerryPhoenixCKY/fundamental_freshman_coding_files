#### 1. 零点的孤立性与唯一性 (Exercise 4.7.6)
*   **(a) 零导数意味着常数**：
    *   如果 $f$ 在 $x_0$ 处的所有导数都为 0，且 $f$ 是解析的（即泰勒级数收敛到它自己），那么 $f(x)$ 在该邻域内恒等于 $f(x_0)$。
    *   *对比 P10 的反例 $e^{-1/x^2}$：那个函数在 0 点所有导数也是 0，但它不是解析函数（它的泰勒级数不等于它自己），所以它不是常数。*

*   **(b) 局部为零意味着全局为零**：
    *   如果一个解析函数 $f$ 在某个小区间 $J$ 上恒等于 0，那么它在整个定义域 $I$ 上都等于 0。
    *   *证明逻辑（P29-P31）*：利用反证法。假设它在某个边界 $b$ 之后不为零，但在 $b$ 点之前都为零。因为解析函数无限光滑，在 $b$ 点的左右极限会导致 $f^{(n)}(b)=0$ 对所有 $n$ 成立。根据 (a)，这意味着 $f$ 在 $b$ 点附近也必须为零，这与 $b$ 是边界的假设矛盾。

*   **(c) 序列零点意味着全局为零**：
    *   如果解析函数 $f$ 有一列零点 $x_n$ 收敛到一个内点 $x_0$，那么 $f$ 恒等于 0。
    *   *证明逻辑（P31-P32）*：利用罗尔定理（Mean Value Theorem）。在相邻的两个零点之间，导数 $f'$ 必有一个零点。这样就构造出一列 $f'$ 的零点也收敛到 $x_0$。同理，$f''$ 也有一列零点收敛到 $x_0$，以此类推。最终得出 $f^{(n)}(x_0)=0$ 对所有 $n$ 成立。根据 (a)，$f$ 必须恒为零。

#### 2. 总结与哲学意义 (P33-P34)
*   **孤立性 (Isolated)**：解析函数如果不恒为零，它的零点必须是**孤立**的（每个零点周围都有一个没有其他零点的小邻域）。
*   **对比**：这再次与 P10-P11 的反例 $e^{-1/x^2}$ 形成鲜明对比。那个非解析函数在 $x<0$ 时恒为 0，但在 $x>0$ 时不为 0，它违反了这个“唯一性原理”。这说明解析函数“刚性”很强，局部决定了整体；而普通光滑函数“柔性”很强，可以局部定义而不影响整体。

**First proof.** We give a proof for $x > x_0$. Denote

$$ I_{n} = \frac{1}{n !} \int_{x_{0}}^{x} ( x - t )^{n} f^{( n + 1 )} ( t ) d t . $$

Integrating by parts we have

$$ \begin{aligned} I_{n} &= \frac{1}{n !}\int_{x_{0}}^{x}(x - t)^{n}d f^{(n)}(t) \\ &= \frac{1}{n !} ( x - t )^{n} f^{( n )} ( t ) \Big |_{x_{0}}^{x} + \frac{1}{( n - 1 ) !} \int_{x_{0}}^{x} ( x - t )^{n - 1} f^{( n )} ( t ) d t \\ &= - \frac{f^{( n )} ( x_{0} )}{n !} ( x - x_{0} )^{n} + I_{n - 1} . \end{aligned} $$

Similarly

$$ I_{n - 1} = - \frac{f^{( n - 1 )} ( x_{0} )}{( n - 1 ) !} ( x - x_{0} )^{n - 1} + I_{n - 2} . $$

Hence

$$ \begin{aligned} I_{n} &= - \frac{f^{( n )} ( x_{0} )}{n !} ( x - x_{0} )^{n} - \frac{f^{( n - 1 )} \big ( x_{0} )}{( n - 1 ) !} \big ( x - x_{0} )^{n - 1} + I_{n - 2} . \end{aligned} $$


By induction we find

$$ \begin{aligned} I_{n} &= - \sum_{k = 1}^{n}{\frac{f^{( k )} ( x_{0} )}{k !}} ( x - x_{0} )^{k} + I_{0} \\ &= - \sum_{k = 1}^{n}{\frac{f^{( k )} ( x_{0} )}{k !} ( x - x_{0} )^{k} + \int_{x_{0}}^{x} f^{\prime} ( t ) d t \\ &= - \sum_{k = 1}^{n}{\frac{f^{( k )} ( x_{0} )}{k !} ( x - x_{0} )^{k} + [ f ( x ) - f ( x_{0} ) ] \\ &= f(x) - P_{n}(x) = R_{n}(x) , \end{aligned} $$

where

$$ P_{n} ( x ) = \sum_{k = 0}^{n}{\frac{f^{( k )} ( x_{0} )}{k !}} ( x - x_{0} )^{k} . $$

**Hence** $R_{n} ( x ) = I_{n}$.
