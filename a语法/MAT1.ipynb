## **第一部分：向量值函数与空间曲线 (Lecture 13-14)**

### **1.1 向量值函数的基本概念**
| 知识点 | 详细内容 |
| :--- | :--- |
| 向量值函数定义 | $\mathbf{r}(t) = f(t)\mathbf{i} + g(t)\mathbf{j} + h(t)\mathbf{k}$ |
| 分量函数 | $f(t), g(t), h(t)$ 为位置向量的分量 |
| 定义域 | 各分量函数定义域的交集 |
| 标量函数 vs 向量函数 | 实值函数称为标量函数，以区别于向量函数 |

### **1.2 极限与连续性**
| 知识点 | 详细内容 |
| :--- | :--- |
| 向量函数极限定义 | $\lim_{t \to t_0} \mathbf{r}(t) = \mathbf{L}$ 的 $\varepsilon$-$\delta$ 定义 |
| 极限的分量判定 | 极限存在当且仅当各分量极限存在 |
| 连续性定义 | $\lim_{t \to t_0} \mathbf{r}(t) = \mathbf{r}(t_0)$ |
| 连续性判定 | 向量函数连续当且仅当各分量函数连续 |

### **1.3 导数与运动学**
| 知识点 | 详细内容 |
| :--- | :--- |
| 导数定义 | $\mathbf{r}'(t) = \lim_{\Delta t \to 0} \frac{\mathbf{r}(t+\Delta t) - \mathbf{r}(t)}{\Delta t}$ |
| 速度向量 | $\mathbf{v}(t) = \frac{d\mathbf{r}}{dt}$，切于曲线 |
| 速率/速度大小 | $\text{Speed} = \|\mathbf{v}(t)\|$ |
| 加速度向量 | $\mathbf{a}(t) = \frac{d\mathbf{v}}{dt} = \frac{d^2\mathbf{r}}{dt^2}$ |
| 运动方向 | $\frac{\mathbf{v}}{\|\mathbf{v}\|}$ 表示时刻 $t$ 的运动方向 |
| 光滑曲线 | $\frac{d\mathbf{r}}{dt}$ 连续且永不为零 |
| 分段光滑曲线 | 由有限条光滑曲线连续拼接而成 |

### **1.4 求导法则**
| 法则 | 公式 |
| :--- | :--- |
| 常数函数法则 | $\frac{d}{dt}\mathbf{C} = \mathbf{0}$ |
| 数乘法则 | $\frac{d}{dt}(c\mathbf{u}) = c\mathbf{u}'$ |
| 函数乘积法则 | $\frac{d}{dt}[f(t)\mathbf{u}(t)] = f'(t)\mathbf{u}(t) + f(t)\mathbf{u}'(t)$ |
| 和/差法则 | $\frac{d}{dt}[\mathbf{u} \pm \mathbf{v}] = \mathbf{u}' \pm \mathbf{v}'$ |
| 点积法则 | $\frac{d}{dt}[\mathbf{u} \cdot \mathbf{v}] = \mathbf{u}' \cdot \mathbf{v} + \mathbf{u} \cdot \mathbf{v}'$ |
| 叉积法则 | $\frac{d}{dt}[\mathbf{u} \times \mathbf{v}] = \mathbf{u}' \times \mathbf{v} + \mathbf{u} \times \mathbf{v}'$ |
| 链式法则 | $\frac{d}{dt}[\mathbf{u}(f(t))] = f'(t)\mathbf{u}'(f(t))$ |
| 常长度向量性质 | 若 $\|\mathbf{r}(t)\|$ 为常数，则 $\mathbf{r} \cdot \frac{d\mathbf{r}}{dt} = 0$ |

### **1.5 积分**
| 知识点 | 详细内容 |
| :--- | :--- |
| 原函数定义 | $\mathbf{R}'(t) = \mathbf{r}(t)$ |
| 不定积分 | $\int \mathbf{r}(t) dt = \mathbf{R}(t) + \mathbf{C}$ |
| 定积分 | $\int_a^b \mathbf{r}(t) dt = \left(\int_a^b f dt\right)\mathbf{i} + \left(\int_a^b g dt\right)\mathbf{j} + \left(\int_a^b h dt\right)\mathbf{k}$ |
| 微积分基本定理 | $\int_a^b \mathbf{r}(t) dt = \mathbf{R}(b) - \mathbf{R}(a)$ |

### **1.6 弧长与参数化**
| 知识点 | 详细内容 |
| :--- | :--- |
| 弧长公式 | $L = \int_a^b \|\mathbf{v}(t)\| dt = \int_a^b \sqrt{(x')^2 + (y')^2 + (z')^2} dt$ |
| 弧长参数 | $s(t) = \int_{t_0}^t \|\mathbf{v}(\tau)\| d\tau$ |
| 弧长参数化 | 可重新参数化曲线为 $\mathbf{r} = \mathbf{r}(t(s))$ |
| $\frac{ds}{dt}$ 的意义 | $\frac{ds}{dt} = \|\mathbf{v}(t)\| > 0$ |

### **1.7 切向量与法平面**
| 知识点 | 详细内容 |
| :--- | :--- |
| 单位切向量 | $\mathbf{T}(t) = \frac{\mathbf{v}(t)}{\|\mathbf{v}(t)\|} = \frac{\mathbf{r}'(t)}{\|\mathbf{r}'(t)\|}$ |
| 切向量公式 | $\mathbf{T} = \frac{d\mathbf{r}}{ds}$ (弧长参数下) |
| 切线方程 | $\mathbf{r} = \mathbf{r}(t_0) + \tau \mathbf{r}'(t_0)$ |
| 法平面方程 | $(\mathbf{r} - \mathbf{r}(t_0)) \cdot \mathbf{r}'(t_0) = 0$ |

### **1.8 曲率与主法向量**
| 知识点 | 详细内容 |
| :--- | :--- |
| 曲率定义 (弧长参数) | $\kappa = \left\|\frac{d\mathbf{T}}{ds}\right\|$ |
| 曲率公式 (一般参数) | $\kappa(t) = \frac{\|\mathbf{T}'(t)\|}{\|\mathbf{v}(t)\|}$ |
| 曲率实用公式 | $\kappa = \frac{\|\mathbf{r}' \times \mathbf{r}''\|}{\|\mathbf{r}'\|^3}$ |
| 主法向量定义 | $\mathbf{N} = \frac{1}{\kappa}\frac{d\mathbf{T}}{ds}$ ($\kappa \neq 0$ 时) |
| 主法向量公式 | $\mathbf{N}(t) = \frac{d\mathbf{T}/dt}{\|d\mathbf{T}/dt\|}$ |
| 曲率几何意义 | 描述曲线弯曲程度，$\kappa$ 越大弯曲越剧烈 |
| 圆的曲率 | $\kappa = \frac{1}{R}$ |

### **1.9 密切圆与曲率半径**
| 知识点 | 详细内容 |
| :--- | :--- |
| 曲率半径 | $\rho = \frac{1}{\kappa}$ |
| 密切圆定义 | 在曲线所在平面内，半径为 $\rho$，圆心在 $\mathbf{r}(t_0) + \rho\mathbf{N}(t_0)$ 的圆 |
| 密切圆性质 | 1. 与曲线相切 2. 有相同曲率 3. 圆心在凹侧 |

### **1.10 挠率与Frenet标架**
| 知识点 | 详细内容 |
| :--- | :--- |
| 副法向量 | $\mathbf{B} = \mathbf{T} \times \mathbf{N}$ |
| Frenet标架 | $\{\mathbf{T}, \mathbf{N}, \mathbf{B}\}$ 构成右手移动标架 |
| 挠率定义 | $\tau = -\frac{d\mathbf{B}}{ds} \cdot \mathbf{N}$ |
| 挠率公式 | $\tau = \frac{(\mathbf{r}' \times \mathbf{r}'') \cdot \mathbf{r}'''}{\|\mathbf{r}' \times \mathbf{r}''\|^2}$ |
| 挠率几何意义 | 描述曲线扭曲程度，可正可负 |
| 曲率 vs 挠率 | 曲率 $\kappa \geq 0$ 衡量弯曲；挠率 $\tau$ 可变号衡量扭曲 |
| Frenet公式 | $\frac{d}{ds}\begin{pmatrix}\mathbf{T} \\ \mathbf{N} \\ \mathbf{B}\end{pmatrix} = \begin{pmatrix}0 & \kappa & 0 \\ -\kappa & 0 & \tau \\ 0 & -\tau & 0\end{pmatrix}\begin{pmatrix}\mathbf{T} \\ \mathbf{N} \\ \mathbf{B}\end{pmatrix}$ |
| 螺旋线特征 | 曲线是螺旋线当且仅当 $\kappa > 0$ 常数且 $\tau \neq 0$ 常数 |

### **1.11 三个特征平面**
| 平面 | 由...确定 | 几何意义 |
| :--- | :--- | :--- |
| 密切平面 | $\mathbf{T}, \mathbf{N}$ | 曲线局部最贴近的平面 |
| 法平面 | $\mathbf{N}, \mathbf{B}$ | 垂直于切线的平面 |
| 从切平面 | $\mathbf{T}, \mathbf{B}$ | 包含切线和副法线的平面 |

### **1.12 加速度的分解**
| 知识点 | 详细内容 |
| :--- | :--- |
| 加速度分解 | $\mathbf{a} = a_T\mathbf{T} + a_N\mathbf{N}$ |
| 切向分量 | $a_T = \frac{d^2s}{dt^2} = \frac{d}{dt}\|\mathbf{v}\|$ (速率变化率) |
| 法向分量 | $a_N = \kappa\left(\frac{ds}{dt}\right)^2 = \kappa\|\mathbf{v}\|^2$ (方向变化率) |
| 加速度位置 | 加速度始终位于密切平面内 |
| 法向加速度公式 | $a_N = \sqrt{\|\mathbf{a}\|^2 - a_T^2}$ |

## **第二部分：多元函数的极限与连续性 (Lecture 15-16)**

### **2.1 平面集合的基本概念**
| 知识点 | 详细内容 |
| :--- | :--- |
| $\delta$-邻域 | $B(x_0, y_0, \delta) = \{(x,y): (x-x_0)^2 + (y-y_0)^2 < \delta^2\}$ |
| $\delta$-方邻域 | $R(x_0, y_0, \delta) = \{(x,y): |x-x_0| < \delta, |y-y_0| < \delta\}$ |
| 内点 | 存在 $\delta > 0$ 使 $B(x_0, y_0, \delta) \subset D$ |
| 边界点 | 任意 $\delta$-邻域同时包含 $D$ 内和 $D$ 外的点 |
| 外点 | 存在 $\delta > 0$ 使 $B(x_0, y_0, \delta) \subset D^c$ |
| 内部 | $D^\circ$，所有内点的集合 |
| 边界 | $\partial D$，所有边界点的集合 |
| 闭包 | $\bar{D} = D^\circ \cup \partial D$ |
| 聚点 | 任意开邻域包含 $D$ 中其他点 |
| 孤立点 | 属于 $D$ 但存在邻域不含 $D$ 中其他点 |

### **2.2 集合的分类**
| 类型 | 定义 |
| :--- | :--- |
| 开集 | 全部由内点组成 ($D^\circ = D$) |
| 闭集 | 包含所有聚点 ($\bar{D} = D$) |
| 有界集 | 包含在某个有限半径的圆盘内 |
| 无界集 | 不是有界集 |
| 连通集 | 任意两点可用 $D$ 内的连续路径连接 |
| 区域 (Domain) | 开集且连通 |
| 闭区域 | 区域的闭包 |

### **2.3 点列与收敛**
| 知识点 | 详细内容 |
| :--- | :--- |
| 点列收敛定义 | $\lim_{n \to \infty} P_n = P \iff \text{dist}(P_n, P) < \varepsilon$ 当 $n > N$ |
| 分量收敛判定 | $P_n \to P \iff x_n \to x$ 且 $y_n \to y$ |
| 有界点列性质 | 有界点列必有收敛子列 (Bolzano-Weierstrass) |
| 闭集性质 | 有界闭集中任意点列有收敛子列且极限在集合内 |
| Cauchy 准则 | 点列收敛当且仅当为 Cauchy 列 |

### **2.4 距离概念**
| 知识点 | 详细内容 |
| :--- | :--- |
| 两点距离 | $\text{dist}(P_1, P_2) = \sqrt{(x_1-x_2)^2 + (y_1-y_2)^2}$ |
| 点到集合距离 | $\text{dist}(P, A) = \inf_{Q \in A} \text{dist}(P, Q)$ |
| 集合间距离 | $\text{dist}(A, B) = \inf_{P \in A, Q \in B} \text{dist}(P, Q)$ |
| 距离公理 | 非负性、对称性、三角不等式 |

### **2.5 多元函数定义**
| 知识点 | 详细内容 |
| :--- | :--- |
| 函数定义 | $w = f(x_1, x_2, \cdots, x_n)$ 将 $n$ 元组映射到实数 |
| 定义域 | 自变量取值的集合 $D$ |
| 值域 | 函数值的集合 |
| 水平集/水平曲线 | $f(x, y) = c$ 的点集 |
| 水平面 | $f(x, y, z) = c$ 的点集 |
| 函数图像 | $\{(x, y, f(x, y))\}$ 构成的曲面 $z = f(x, y)$ |

### **2.6 二重极限**
| 知识点 | 详细内容 |
| :--- | :--- |
| 极限定义 | $\lim_{(x,y) \to (x_0,y_0)} f(x,y) = L$ 的 $\varepsilon$-$\delta$ 定义 |
| 极限存在条件 | 任意方向趋近必须得到相同极限值 |
| 极限运算法则 | 和、差、积、商、幂、根号法则 |
| 双路径检验法 | 沿两条不同路径极限不同则极限不存在 |
| 累次极限 vs 二重极限 | 累次极限存在且相等不能推出二重极限存在 |
| 重要结论 | 沿所有直线极限相同不能保证极限存在 |

### **2.7 连续性**
| 知识点 | 详细内容 |
| :--- | :--- |
| 连续定义 | 1. 函数在该点有定义 2. 极限存在 3. 极限等于函数值 |
| 连续函数运算 | 连续函数的和、差、积、商、幂、复合仍连续 |
| 多项式与有理函数 | 在定义域内处处连续 |
| 复合函数连续性 | 若 $f$ 在 $(x_0,y_0)$ 连续，$g$ 在 $f(x_0,y_0)$ 连续，则 $g \circ f$ 连续 |

### **2.8 一致连续性**
| 知识点 | 详细内容 |
| :--- | :--- |
| 一致连续定义 | $\delta$ 只依赖于 $\varepsilon$，与点的位置无关 |
| 重要定理 | 有界闭集上的连续函数必一致连续 |
| 一致收敛 | $\lim_{x \to x_0} f(x,y) = g(y)$ 对 $y$ 一致成立 |

### **2.9 最值定理**
| 知识点 | 详细内容 |
| :--- | :--- |
| 极值定理 | 有界闭集上的连续函数必取得绝对最大值和最小值 |
| 证明方法 | 利用确界原理和点列收敛性质 |

### **2.10 三元及以上函数**
| 知识点 | 详细内容 |
| :--- | :--- |
| 极限与连续性 | 定义和性质可推广到三元及更多变量 |
| 空间集合概念 | 内点、边界点、开集、闭集等定义类似 |
| 有界闭集性质 | 空间中有界闭集上的连续函数同样一致连续且取到最值 |

## **第三部分：偏导数与可微性 (Lecture 17)**

### **3.1 偏导数定义**
| 知识点 | 详细内容 |
| :--- | :--- |
| 对 $x$ 偏导数 | $\frac{\partial f}{\partial x}(x_0,y_0) = \lim_{h \to 0} \frac{f(x_0+h, y_0) - f(x_0, y_0)}{h}$ |
| 对 $y$ 偏导数 | $\frac{\partial f}{\partial y}(x_0,y_0) = \lim_{h \to 0} \frac{f(x_0, y_0+h) - f(x_0, y_0)}{h}$ |
| 记号 | $f_x, f_y, \frac{\partial z}{\partial x}, \frac{\partial z}{\partial y}$ 等 |
| 几何意义 | 截面曲线在切点处的切线斜率 |
| 物理意义 | 固定其他变量时函数对某一变量的变化率 |

### **3.2 偏导数与连续性的关系**
| 知识点 | 详细内容 |
| :--- | :--- |
| 重要区别 | 偏导数存在 **不能** 推出函数连续 (与单变量不同) |
| 充分条件 | 若偏导数存在且连续，则函数连续 |
| 反例 | 存在偏导数但不连续的函数 |

### **3.3 高阶偏导数**
| 知识点 | 详细内容 |
| :--- | :--- |
| 二阶偏导数 | $f_{xx}, f_{xy}, f_{yx}, f_{yy}$ |
| 混合偏导数 | $f_{xy} = (f_x)_y$, $f_{yx} = (f_y)_x$ |
| Clairaut 定理 | 若 $f_{xy}, f_{yx}$ 连续，则 $f_{xy} = f_{yx}$ |
| 反例 | 存在 $f_{xy} \neq f_{yx}$ 的函数 (当连续性条件不满足时) |
| 更高阶导数 | 三阶、四阶等，记号如 $f_{xxx}, f_{xxy}$ 等 |
| 求导顺序 | 只要各阶导数连续，求导顺序可交换 |

### **3.4 可微性**
| 知识点 | 详细内容 |
| :--- | :--- |
| 可微定义 1 | $\Delta z = a\Delta x + b\Delta y + \varepsilon_1\Delta x + \varepsilon_2\Delta y$，其中 $\varepsilon_1, \varepsilon_2 \to 0$ |
| 可微定义 2 | $\Delta z = f_x\Delta x + f_y\Delta y + \varepsilon_1\Delta x + \varepsilon_2\Delta y$ |
| 等价形式 | $\Delta z = f_x\Delta x + f_y\Delta y + \varepsilon\sqrt{(\Delta x)^2 + (\Delta y)^2}$ |
| 可微与偏导 | 可微 $\Rightarrow$ 偏导存在，但反之不成立 |
| 可微与连续 | 可微 $\Rightarrow$ 连续 |
| 连续可微 | 偏导数存在且连续 $\Rightarrow$ 可微 (充分非必要条件) |
| 反例 | 存在可微但偏导数不连续的函数 |

### **3.5 全微分**
| 知识点 | 详细内容 |
| :--- | :--- |
| 全微分定义 | $df = f_x dx + f_y dy$，其中 $dx = \Delta x, dy = \Delta y$ |
| 三元函数 | $df = f_x dx + f_y dy + f_z dz$ |
| 几何意义 | 切平面上的增量近似 |

### **3.6 可微性层次关系**
```
连续可微 (偏导数连续)
    ⇓
可微性
    ⇓
有偏导数  且  连续
```

## **第四部分：链式法则与梯度 (Lecture 18)**

### **4.1 链式法则**
| 情形 | 公式 |
| :--- | :--- |
| 单变量中间 | $w = f(x(t), y(t))$, $\frac{dw}{dt} = \frac{\partial f}{\partial x}\frac{dx}{dt} + \frac{\partial f}{\partial y}\frac{dy}{dt}$ |
| 三变量单参数 | $w = f(x(t), y(t), z(t))$, $\frac{dw}{dt} = f_x\frac{dx}{dt} + f_y\frac{dy}{dt} + f_z\frac{dz}{dt}$ |
| 两中间变量 | $w = f(x(r,s), y(r,s))$, $\frac{\partial w}{\partial r} = \frac{\partial w}{\partial x}\frac{\partial x}{\partial r} + \frac{\partial w}{\partial y}\frac{\partial y}{\partial r}$ |
| 一般形式 | 对每个自变量求偏导，乘以中间变量对该自变量的偏导，再求和 |
| 使用条件 | 要求函数 **可微**，仅偏导存在不够 |

### **4.2 隐函数求导**
| 情形 | 公式 |
| :--- | :--- |
| 两变量隐函数 | $F(x,y)=0$ 定义 $y=f(x)$, $\frac{dy}{dx} = -\frac{F_x}{F_y}$ ($F_y \neq 0$) |
| 三变量隐函数 | $F(x,y,z)=0$ 定义 $z=f(x,y)$, $\frac{\partial z}{\partial x} = -\frac{F_x}{F_z}$, $\frac{\partial z}{\partial y} = -\frac{F_y}{F_z}$ |
| 推导方法 | 对方程两边用链式法则求导 |

### **4.3 方向导数**
| 知识点 | 详细内容 |
| :--- | :--- |
| 定义 | $D_{\mathbf{u}}f(P_0) = \lim_{s \to 0^+} \frac{f(P_0 + s\mathbf{u}) - f(P_0)}{s}$ |
| 单位向量 | $\mathbf{u} = u_1\mathbf{i} + u_2\mathbf{j}$, $u_1^2 + u_2^2 = 1$ |
| 几何意义 | 曲面沿方向 $\mathbf{u}$ 的截面曲线切线斜率 |
| 物理意义 | 物理量 (如温度) 沿某方向的瞬时变化率 |
| 偏导数是特例 | $f_x, f_y$ 分别是沿 $\mathbf{i}, \mathbf{j}$ 方向的方向导数 |

### **4.4 梯度**
| 知识点 | 详细内容 |
| :--- | :--- |
| 梯度定义 | $\nabla f = \frac{\partial f}{\partial x}\mathbf{i} + \frac{\partial f}{\partial y}\mathbf{j}$ (二元) |
| 三元梯度 | $\nabla f = \frac{\partial f}{\partial x}\mathbf{i} + \frac{\partial f}{\partial y}\mathbf{j} + \frac{\partial f}{\partial z}\mathbf{k}$ |
| 核心公式 | $D_{\mathbf{u}}f = \nabla f \cdot \mathbf{u}$ (要求 $f$ 可微) |
| 角度形式 | $D_{\mathbf{u}}f = \|\nabla f\| \cos\theta$ |

### **4.5 梯度的性质**
| 性质 | 详细内容 |
| :--- | :--- |
| 最大增长方向 | 沿 $\nabla f$ 方向，方向导数最大值为 $\|\nabla f\|$ |
| 最大下降方向 | 沿 $-\nabla f$ 方向，方向导数最小值为 $-\|\nabla f\|$ |
| 零变化方向 | 与 $\nabla f$ 垂直的方向，方向导数为 0 |
| 与水平集关系 | 梯度垂直于水平曲线/水平面 |
| 切线方程 | $f_x(x_0,y_0)(x-x_0) + f_y(x_0,y_0)(y-y_0) = 0$ |

### **4.6 梯度的运算法则**
| 法则 | 公式 |
| :--- | :--- |
| 和法则 | $\nabla(f+g) = \nabla f + \nabla g$ |
| 差法则 | $\nabla(f-g) = \nabla f - \nabla g$ |
| 数乘法则 | $\nabla(kf) = k\nabla f$ |
| 积法则 | $\nabla(fg) = f\nabla g + g\nabla f$ |
| 商法则 | $\nabla(\frac{f}{g}) = \frac{g\nabla f - f\nabla g}{g^2}$ |

### **4.7 沿路径的导数**
| 知识点 | 详细内容 |
| :--- | :--- |
| 公式 | $\frac{d}{dt}f(\mathbf{r}(t)) = \nabla f(\mathbf{r}(t)) \cdot \mathbf{r}'(t)$ |
| 应用 | 计算函数沿曲线的变化率 |

## **第五部分：几何应用 (Lecture 18 第二部分)**

### **5.1 切平面与法线 (显式曲面)**
| 知识点 | 详细内容 |
| :--- | :--- |
| 曲面方程 | $S: z = f(x, y)$ |
| 切向量 1 | $\mathbf{u}_1 = \mathbf{i} + f_x\mathbf{k}$ (沿 $x$ 方向) |
| 切向量 2 | $\mathbf{u}_2 = \mathbf{j} + f_y\mathbf{k}$ (沿 $y$ 方向) |
| 法向量 | $\mathbf{n} = f_x\mathbf{i} + f_y\mathbf{j} - \mathbf{k}$ |
| 切平面方程 | $f_x(x_0,y_0)(x-x_0) + f_y(x_0,y_0)(y-y_0) = z - z_0$ |
| 法线方程 | $\frac{x-x_0}{f_x} = \frac{y-y_0}{f_y} = \frac{z-z_0}{-1}$ |
| 可微性要求 | 要求 $f$ 可微，仅偏导存在不足以保证切平面存在 |

### **5.2 切平面与法线 (隐式曲面)**
| 知识点 | 详细内容 |
| :--- | :--- |
| 曲面方程 | $S: F(x, y, z) = c$ |
| 法向量 | $\nabla F(P)$ |
| 切平面方程 | $F_x(P)(x-x_0) + F_y(P)(y-y_0) + F_z(P)(z-z_0) = 0$ |
| 法线方程 | $\frac{x-x_0}{F_x} = \frac{y-y_0}{F_y} = \frac{z-z_0}{F_z}$ |
| 两曲面交线切线 | 方向为 $\nabla F \times \nabla G$ |

### **5.3 线性化**
| 知识点 | 详细内容 |
| :--- | :--- |
| 线性化定义 | $L(x,y) = f(x_0,y_0) + f_x(x_0,y_0)(x-x_0) + f_y(x_0,y_0)(y-y_0)$ |
| 标准线性近似 | $f(x,y) \approx L(x,y)$ |
| 误差估计 | $|E(x,y)| \leq \frac{M}{2}(|x-x_0| + |y-y_0|)^2$ |
| 几何意义 | 切平面近似，误差是高阶小量 |
| 唯一性 | 满足切触性质的平面唯一，即为切平面 |
| 三元线性化 | $L(x,y,z) = f(P_0) + f_x\Delta x + f_y\Delta y + f_z\Delta z$ |

## **第六部分：极值问题 (Lecture 20)**

### **6.1 局部极值定义**
| 知识点 | 详细内容 |
| :--- | :--- |
| 局部极大值 | $f(x_0,y_0) \geq f(x,y)$ 在某邻域内 |
| 局部极小值 | $f(x_0,y_0) \leq f(x,y)$ 在某邻域内 |
| 临界点 | $f_x = f_y = 0$ 或偏导不存在的点 |
| 鞍点 | 临界点但非极值点，任意邻域内既有更大值也有更小值 |

### **6.2 一阶导数检验**
| 知识点 | 详细内容 |
| :--- | :--- |
| 定理 | 若 $f$ 在内点取局部极值且偏导存在，则 $f_x = f_y = 0$ |
| 几何意义 | 极值点处切平面水平 |
| 极值位置 | 只可能在临界点或边界点取得 |

### **6.3 二阶导数检验**
| 判别式 $D$ | $f_{xx}$ | 结论 |
| :--- | :--- | :--- |
| $D > 0$ | $f_{xx} < 0$ | 局部极大值 (负定) |
| $D > 0$ | $f_{xx} > 0$ | 局部极小值 (正定) |
| $D < 0$ | 任意 | 鞍点 (不定) |
| $D = 0$ | 任意 | 检验失效，需其他方法 |
| 判别式定义 | $D = f_{xx}f_{yy} - f_{xy}^2 = \det(D^2f)$ |
| Hessian 矩阵 | $D^2f = \begin{pmatrix}f_{xx} & f_{xy} \\ f_{yx} & f_{yy}\end{pmatrix}$ |

### **6.4 绝对极值求法**
| 步骤 | 内容 |
| :--- | :--- |
| 步骤 1 | 列出所有内部临界点并计算函数值 |
| 步骤 2 | 找出边界上的局部极值点并计算函数值 |
| 步骤 3 | 比较所有值，最大为绝对最大，最小为绝对最小 |

### **6.5 拉格朗日乘数法 (单约束)**
| 知识点 | 详细内容 |
| :--- | :--- |
| 问题形式 | 优化 $f(x,y,z)$ 受约束 $g(x,y,z) = 0$ |
| 必要条件 | $\nabla f = \lambda \nabla g$, $g = 0$ |
| 几何意义 | 目标函数与约束函数的梯度平行 |
| 方程组 | $\begin{cases}f_x = \lambda g_x \\ f_y = \lambda g_y \\ f_z = \lambda g_z \\ g = 0\end{cases}$ |
| 解的性质 | 解是可能的极值点，需进一步验证 |

### **6.6 拉格朗日乘数法 (双约束)**
| 知识点 | 详细内容 |
| :--- | :--- |
| 问题形式 | 优化 $f$ 受约束 $g_1 = 0, g_2 = 0$ |
| 必要条件 | $\nabla f = \lambda \nabla g_1 + \mu \nabla g_2$ |
| 方程组 | $\begin{cases}\nabla f = \lambda \nabla g_1 + \mu \nabla g_2 \\ g_1 = 0 \\ g_2 = 0\end{cases}$ |
| 几何意义 | $\nabla f$ 在 $\nabla g_1, \nabla g_2$ 张成的平面内 |
| 应用示例 | 求椭圆上距原点最近/最远点 |

### **6.7 正交梯度定理**
| 知识点 | 详细内容 |
| :--- | :--- |
| 定理内容 | 若 $f$ 在曲线 $C$ 上 $P_0$ 点取条件极值，则 $\nabla f \perp C$ 于 $P_0$ |
| 证明思路 | $\frac{d}{dt}f(\mathbf{r}(t)) = \nabla f \cdot \mathbf{r}' = 0$ |
| 推论 | 条件极值点处 $\nabla f = \lambda \nabla g$ |

### **6.8 约束变量下的偏导数**
| 知识点 | 详细内容 |
| :--- | :--- |
| 问题 | $w = f(x,y,z)$ 且 $g(x,y,z) = 0$，求 $\frac{\partial w}{\partial x}$ |
| 方法 1 | 选 $x,y$ 为自变量：$\frac{\partial w}{\partial x} = f_x + f_z\frac{\partial z}{\partial x} = f_x - f_z\frac{g_x}{g_z}$ |
| 方法 2 | 选 $x,z$ 为自变量：$\frac{\partial w}{\partial x} = f_x + f_y\frac{\partial y}{\partial x} = f_x - f_y\frac{g_x}{g_y}$ |
| 关键 | 明确哪些是独立变量，哪些是依赖变量 |

## **第七部分：泰勒公式 (Lecture 20)**

### **7.1 二阶导数检验的证明**
| 知识点 | 详细内容 |
| :--- | :--- |
| 证明工具 | 泰勒公式带 Lagrange 余项 |
| 参数化 | $F(t) = f(x_0+th, y_0+tk)$ |
| 二次型 | $Q = f_{xx}h^2 + 2f_{xy}hk + f_{yy}k^2$ |
| 配方分析 | $Q = \frac{1}{f_{xx}}[(f_{xx}h + f_{xy}k)^2 + (f_{xx}f_{yy}-f_{xy}^2)k^2]$ |

### **7.2 二元泰勒公式**
| 知识点 | 详细内容 |
| :--- | :--- |
| $n$ 阶公式 | $f(x_0+h, y_0+k) = f(x_0,y_0) + (hf_x + kf_y) + \frac{1}{2!}(h^2f_{xx} + 2hkf_{xy} + k^2f_{yy}) + \cdots + \frac{1}{n!}(h\frac{\partial}{\partial x} + k\frac{\partial}{\partial y})^nf + R_n$ |
| 余项 | $R_n = \frac{1}{(n+1)!}(h\frac{\partial}{\partial x} + k\frac{\partial}{\partial y})^{n+1}f|_{(x_0+ch, y_0+ck)}$ |
| 原点展开 | $x_0=y_0=0$ 时的特殊形式 |
| 求和形式 | $f(x,y) = f(x_0,y_0) + \sum_{j=1}^n \frac{1}{j!}\sum_{i=0}^j \frac{\partial^j f}{\partial x^i \partial y^{j-i}}(x-x_0)^i(y-y_0)^{j-i} + R_n$ |

### **7.3 应用示例**
| 应用 | 详细内容 |
| :--- | :--- |

| 示例 | $\sin x \sin y \approx xy$ 在原点附近 |
