
### ✅ **第 1 步：明确「输入规模」\( n \) 是什么？**

> 📌 **关键问题**：这段代码处理的数据量由哪个变量决定？

| 场景 | 输入规模 \( n \) |
|------|----------------|
| 函数参数是 `arr`（列表） | \( n = \text{len(arr)} \) |
| 函数参数是 `s`（字符串） | \( n = \text{len(s)} \) |
| 处理二维矩阵 `matrix` | \( n = \text{行数} \times \text{列数} \)，或单独用 \( m, n \) |
| 递归处理树 | \( n = \text{节点总数} \) |
| 数值算法（如求 n!） | \( n = \text{数值本身} \) |

> ⚠️ 如果有多个输入（如两个数组），通常取**主导项**，或用 \( O(f(m, n)) \) 表示。

---

### ✅ **第 2 步：识别「主导控制结构」**

找出决定执行次数的核心结构：

| 结构 | 特征 | 初步复杂度猜测 |
|------|------|--------------|
| **单层循环** | `for i in range(n)` | \( O(n) \) |
| **嵌套循环** | 循环内还有循环 | \( O(n^k) \)，k=层数 |
| **递归** | 函数调用自身 | 需写递推式 |
| **分治/二分** | 每次问题规模减半 | \( O(\log n) \) 或 \( O(n \log n) \) |
| **无循环/递归** | 顺序语句 | \( O(1) \) |

> 🔍 **注意**：有些循环看起来嵌套，但内层次数固定（如 `for j in range(10)`）→ 实际是 \( O(n) \)。

---

### ✅ **第 3 步：逐行估算「基本操作」执行次数**

根据讲义 p.5，**基本操作**包括：
- 赋值、算术运算（`+`, `*`）
- 比较（`==`, `<`）
- 数组/列表索引（`arr[i]`）
- 函数调用（不包括函数内部）

对每行代码，问：**在最坏情况下，这行最多执行多少次？（用 \( n \) 表示）**

#### 🧪 示例：
```python
def example(arr):
    n = len(arr)                 # 1 次
    total = 0                    # 1 次
    for i in range(n):           # 执行 n 次
        for j in range(i+1):     # 执行 1+2+...+n = n(n+1)/2 ≈ n²/2 次
            total += arr[i] * arr[j]  # 每次 3 次操作（索引×2 + 乘 + 加）
    return total                 # 1 次
```
- 总操作数 ≈ \( 1 + 1 + n + \frac{n^2}{2} \times 3 + 1 = \frac{3}{2}n^2 + n + 3 \)

---

### ✅ **第 4 步：写出总操作数表达式，并简化为大 O**

#### 简化规则（来自讲义 p.9–10）：
1. **忽略常数因子**：\( 5n^2 \rightarrow n^2 \)
2. **忽略低阶项**：\( n^2 + n + 1 \rightarrow n^2 \)
3. **保留最高增长阶**

> ✅ 上例：\( \frac{3}{2}n^2 + n + 3 \rightarrow O(n^2) \)

#### 常见增长阶排序（讲义 p.11）：
\[
O(1) < O(\log n) < O(n) < O(n \log n) < O(n^2) < O(n^3) < O(2^n) < O(n!)
\]

### ✅ **第 5 步：特殊结构处理技巧**

#### A. **递归函数** → 写递推关系式
```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)
```
- 递推式：\( T(n) = T(n-1) + T(n-2) + O(1) \)
- 解得：\( T(n) = O(2^n) \)（指数级）

> 💡 若递归是**分治型**（如 `T(n) = 2T(n/2) + O(n)`），用**主定理**快速判断。

#### B. **内置函数/方法** → 查其复杂度！
| 操作 | 时间复杂度 |
|------|----------|
| `list.append()` | \( O(1) \)（摊还） |
| `list.insert(0, x)` | \( O(n) \) |
| `x in list` | \( O(n) \) |
| `x in set` / `dict[key]` | \( O(1) \) |
| `sorted(list)` | \( O(n \log n) \) |
| `str1 + str2`（在循环中） | \( O(n^2) \)（应改用 `join`）

> ⚠️ 忽略内置函数复杂度是常见错误！

#### C. **短路/提前退出** → 分析**最坏情况**
```python
for i in range(n):
    if arr[i] == target:  # 最好 O(1)，最坏 O(n)
        return i
```
- 按讲义原则：**关注最坏情况** → \( O(n) \)

## 🧩 举个完整例子

**代码**：
```python
def count_pairs(arr, target):
    seen = set()
    count = 0
    for num in arr:          # n 次
        if target - num in seen:  # O(1)
            count += 1
        seen.add(num)        # O(1)
    return count
```

**分析步骤**：
1. \( n = \text{len(arr)} \)
2. 主导结构：单层循环
3. 每次循环：2 次 \( O(1) \) 操作 → 总操作数 ≈ \( 2n \)
4. 简化：\( O(n) \)
5. 验证：哈希表查找 → 线性时间 → 合理

✅ **结论：时间复杂度为 \( O(n) \)**

## ✅ 总结：你的分析 checklist

当你面对一段新代码时，按顺序问自己：

1. **输入规模 \( n \) 是什么？**
2. **有没有循环？几层？每层跑多少次？**
3. **有没有递归？递推式是什么？**
4. **有没有调用内置函数？它们的复杂度是多少？**
5. **最坏情况下，总操作数表达式是什么？**
6. **简化后，属于哪种增长阶？**

## 🔢 一、排序算法（Sorting Algorithms）

| 算法 | 最好情况 | 平均情况 | 最坏情况 | 空间复杂度 | 是否稳定 | 是否原地 |
|------|--------|--------|--------|----------|--------|--------|
| **冒泡排序** (Bubble Sort) | \( O(n) \) | \( O(n^2) \) | \( O(n^2) \) | \( O(1) \) | ✅ | ✅ |
| **选择排序** (Selection Sort) | \( O(n^2) \) | \( O(n^2) \) | \( O(n^2) \) | \( O(1) \) | ❌ | ✅ |
| **插入排序** (Insertion Sort) | \( O(n) \) | \( O(n^2) \) | \( O(n^2) \) | \( O(1) \) | ✅ | ✅ |
| **希尔排序** (Shell Sort) | \( O(n \log n) \) | \( O(n^{1.3}) \) | \( O(n^2) \) | \( O(1) \) | ❌ | ✅ |
| **归并排序** (Merge Sort) | \( O(n \log n) \) | \( O(n \log n) \) | \( O(n \log n) \) | \( O(n) \) | ✅ | ❌ |
| **快速排序** (Quick Sort) | \( O(n \log n) \) | \( O(n \log n) \) | \( O(n^2) \) | \( O(\log n) \) | ❌ | ✅ |
| **堆排序** (Heap Sort) | \( O(n \log n) \) | \( O(n \log n) \) | \( O(n \log n) \) | \( O(1) \) | ❌ | ✅ |
| **计数排序** (Counting Sort) | \( O(n + k) \) | \( O(n + k) \) | \( O(n + k) \) | \( O(k) \) | ✅ | ❌ |
| **桶排序** (Bucket Sort) | \( O(n) \) | \( O(n) \) | \( O(n^2) \) | \( O(n) \) | ✅ | ❌ |
| **基数排序** (Radix Sort) | \( O(d(n + k)) \) | \( O(d(n + k)) \) | \( O(d(n + k)) \) | \( O(n + k) \) | ✅ | ❌ |

> 📌 注：
> - \( n \)：元素个数；\( k \)：值域大小；\( d \)：位数。
> - **比较排序下限**：任何基于比较的排序算法最坏情况至少 \( \Omega(n \log n) \)（讲义 p.14 提到“高效 vs 暴力”）。
> - **非比较排序**（计数/桶/基数）可突破此限制，但有数据范围要求。

## 🔍 二、搜索算法（Searching Algorithms）

| 算法 | 数据结构 | 时间复杂度（平均/最坏） | 备注 |
|------|--------|----------------------|------|
| **线性搜索** (Linear Search) | 无序数组 | \( O(n) \) | 最坏需遍历全部 |
| **二分搜索** (Binary Search) | **有序数组** | \( O(\log n) \) | 讲义 p.25–26 重点讲解 |
| **哈希表查找** | Hash Table | \( O(1) \) / \( O(n) \) | 平均 \( O(1) \)，最坏冲突多时 \( O(n) \) |
| **二叉搜索树查找** | BST | \( O(\log n) \) / \( O(n) \) | 平衡时 \( O(\log n) \)，退化成链表时 \( O(n) \) |
| **平衡 BST 查找**（如 AVL, Red-Black） | 平衡树 | \( O(\log n) \) | 保证最坏 \( O(\log n) \) |

## 🌳 三、图算法（Graph Algorithms）

设图有 \( V \) 个顶点，\( E \) 条边。

| 算法 | 问题 | 时间复杂度 | 数据结构依赖 |
|------|------|----------|------------|
| **BFS**（广度优先搜索） | 最短路径（无权图） | \( O(V + E) \) | 邻接表 |
| **DFS**（深度优先搜索） | 连通性、拓扑排序 | \( O(V + E) \) | 邻接表 |
| **Dijkstra** | 单源最短路径（非负权） | \( O((V + E) \log V) \) | 优先队列（堆） |
| **Bellman-Ford** | 单源最短路径（可负权） | \( O(VE) \) | — |
| **Floyd-Warshall** | 所有对最短路径 | \( O(V^3) \) | 邻接矩阵 |
| **Kruskal** | 最小生成树 | \( O(E \log E) \) | 并查集 |
| **Prim** | 最小生成树 | \( O(E \log V) \) | 优先队列 |

## ⚠️ 六、重要概念回顾（来自 Lecture 8）

1. **多项式时间 vs 指数时间**（讲义 p.17）
   - **高效算法**：\( O(n^k) \)（k 为常数）
   - **低效算法**：\( O(2^n), O(n!) \) → “即使宇宙毁灭也算不完”

2. **增长阶顺序**（讲义 p.11）
   \[
   O(1) < O(\log n) < O(n) < O(n \log n) < O(n^2) < O(n^3) < \cdots < O(2^n) < O(n!)
   \]

3. **大 O 忽略常数和低阶项**（讲义 p.9）
   - \( 5n^2 + 3n + 100 = O(n^2) \)
